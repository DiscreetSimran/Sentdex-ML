```python
# HORIZONTAL WINNER CALCULATION
# there are 3 different ways of calulating the winner in a tic tac toe game 
# horizontally, vertically, diagonally


game=[[1,1,1],
      [0,2,0],
      [2,2,0 ]]


# one way to calculate winner horizonatally is to check in each row for each element and use a flag to see if they are all equal
# this method can be a little messy
# next method :-

def win(current_game):
    for row in current_game:
        print(row)
        col1=row[0]
        col2=row[1]
        col3=row[2]
        if col1==col2==col3:
            print("winner !!")
            
            
win(game)


# as we have two players :- 1 & 2
# and the first row has all ones that means the player 1 wins but the problem here is 
# if at any time you want to upgrade your code and make a 4x4 tic tac toe game .. you will have to hard code changes in 
# your horizonatal winner calculation like adding "col4=row[3]" in the win()
# this is called technical debt
# which you shouldn't because calculation of a horizontal winner will remain the same
# no matter the size of the tic tac toe game
```

```python
# making the game dynamic
# a basic dynamic idea of doing this would be :-


game=[[1,1,1],
      [0,2,0],
      [2,2,0 ]]

def win(current_game):
    for row in current_game:
        print(row)
        all_match=True
        for item in row:
            print(item,row[0])
            if(item!=row[0]):
                all_match=False
        if all_match:
            print("winner")

win(game)



# we can make this code better 
# maybe "print(item,row[0])" is not needed , it's just for displaying thw working and for debugging purposes 
# but still it's more than enough lines for this problem
```

```python
# so to make the code better use google
# google "check if all items in a list are same in python"
# one of the answers on stack over flow is :-
# "x.count(x[0]) == len(x)"
# this checks the number of occurences of the first element of the list in the list itself 
# and if it's equal to the length of the list that means x[0] is repeated that many times 
# means all elements are equal



game=[[1,1,1],
      [0,2,0],
      [2,2,0 ]]

def win(current_game):
    for row in current_game:
        print(row)
        if ((row.count(row[0])==len(row))==True) and row[0]!=0:
            print("winner", row[0])
            
win(game)



# here we had hard coded game[[]] , but originally it will be all zeros 
```
